The data in this analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
play_store_stat = pd.read_csv('C:/Users/rabia/Desktop/Thinkful/Capstone project 1/googleplaystore.csv')
play_store_reviews = pd.read_csv('C:/Users/rabia/Desktop/Thinkful/Capstone project 1/googleplaystore_user_reviews.csv')
```

```python
play_store_stat.head()
```

```python
play_store_stat['installs_numeric'] = pd.to_numeric(play_store_stat['Installs'].str.replace('+','').str.replace(',',''), errors='ignore')
play_store_stat.head()
```

```python
play_store_reviews.groupby('App').count()
```

```python
len(play_store_reviews)
```

```python
play_store_reviews_grouped = play_store_reviews.groupby('App').mean()
```

```python
play_store_reviews_grouped.head()
```

```python
len(play_store_reviews_grouped)
```

```python
play_store = play_store_stat.merge(play_store_reviews_grouped, on='App')
```

```python
play_store.head()
```

```python
play_store.Category.value_counts().head(10)
```

```python
plt.figure(figsize=(20,10))
play_store.Category.value_counts().plot(kind='bar', color = 'blue')
plt.xticks(fontsize=15, rotation=90)
plt.yticks(fontsize=15, rotation=0)
plt.ylabel('Category Count', fontsize = 20)
```

```python
play_store = play_store.sort_values('Rating', ascending = False).reset_index(drop= True)
plt.figure(figsize=(20,10))
play_store.Rating.value_counts().plot(kind='bar', color = 'red')
plt.xticks(fontsize=15, rotation=0)
plt.ylabel('Rating Count' , fontsize = 20)
plt.yticks(fontsize=15, rotation=0)
```

```python
play_store_category = play_store.groupby('Category').mean()
play_store_category
```

```python
plt.rcParams['figure.figsize']=(20,10)
play_store_category[('Rating')].plot(kind='bar')
plt.xticks(fontsize=15, rotation=90)
plt.ylabel('Average Rating' , fontsize = 20)
plt.yticks(fontsize=15, rotation=0)
```

```python
plt.rcParams['figure.figsize']=(20,10)
play_store_category[('Sentiment_Polarity')].plot(kind='bar')
plt.xticks(fontsize=15, rotation=90)
plt.ylabel('Average Sentiment Polarity' , fontsize = 20)
plt.yticks(fontsize=15, rotation=0)
```

```python
plt.rcParams['figure.figsize']=(20,10)
play_store_category[('Sentiment_Subjectivity')].plot(kind='bar')
plt.xticks(fontsize=15, rotation=90)
plt.ylabel('Average Sentiment Subjectivity' , fontsize = 20)
plt.yticks(fontsize=15, rotation=0)
```

```python
plt.figure(figsize=(20,10))
play_store.Type.value_counts().plot(kind='bar', color = 'green')
plt.xticks(fontsize=20, rotation=0)
plt.yticks(fontsize=15, rotation=0)
plt.ylabel('Type Count', fontsize = 20)
play_store.Type.value_counts()
```

```python
play_store_installs = play_store.groupby('installs_numeric').mean()
play_store_installs
```

```python

plt.rcParams['figure.figsize']=(20,10)
play_store_installs[('Rating')].plot(kind='bar')
plt.xticks(fontsize=15, rotation=90)
plt.ylabel('Average rating' , fontsize = 20)
plt.yticks(fontsize=15, rotation=0)
```
